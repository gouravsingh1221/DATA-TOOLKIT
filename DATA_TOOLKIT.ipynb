#Q2  How does broadcasting work in NumPy ?

Ans.- Broadcasting in NumPy is a feature that allows you to perform operations on arrays of different shapes without writing extra code. Normally, you can only do operations on arrays that are the same size, but broadcasting automatically stretches the smaller array to match the larger one — without actually copying data. For example, if you add a number to a NumPy array, that number is automatically added to every element in the array. Similarly, if you add a smaller array (like a row or column) to a larger one, NumPy adjusts it so the operation works across all elements. Broadcasting makes code shorter, faster, and easier to write when working with arrays of different sizes.

#Q8  What does the term “vectorized operation” mean in NumPy ?

#Q17  Explain the difference between apply() and map() in Pandas ?

Ans.- In Pandas, both apply() and map() are used to apply functions to data, but they are used in different ways. The map() function is used only on a single Series (like one column) and is best for simple functions or value replacements. For example, you can use it to change "Yes" to 1 and "No" to 0 in a column. On the other hand, apply() is more flexible and powerful — it can be used on both Series and DataFrames, and it supports custom functions, even ones that do more complex operations. So, use map() for simple tasks on one column, and use apply() when you need more control or want to work across rows or multiple columns.
